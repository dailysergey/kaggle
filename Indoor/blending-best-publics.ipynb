```python
import pandas as pd
import numpy as np
```

```python
submision_file=pd.read_csv('../input/indoor-location-navigation/sample_submission.csv')
```

```python
submision_file = pd.DataFrame(submision_file['site_path_timestamp'])
submision_file['floor'] = 0
submision_file['x'] = 0
submision_file['y'] = 0
```

```python
first_score = pd.read_csv('../input/villageindoorbestsubmissions/submission_lb5.296.csv')
third_score = pd.read_csv('../input/villageindoorbestsubmissions/submission_lb6196.csv')
second_score = pd.read_csv('../input/villageindoorbestsubmissions/submission_lb7.443_LSTM_Unified_Wi-Fi.csv')
fourth_score = pd.read_csv('../input/villageindoorbestsubmissions/submission_lb4.909.csv')
fifth_score = pd.read_csv('../input/villageindoorbestsubmissions/submission_lb7.274_LSTM.csv')
six_score = pd.read_csv('../input/villageindoorbestsubmissions/submission_lb15.420_MLP.csv')
```

```python
submision_file['first_floor'] = first_score.floor
submision_file['first_x'] = first_score.x
submision_file['first_y'] = first_score.y
```

```python
submision_file['second_floor'] = second_score.floor
submision_file['second_x'] = second_score.x
submision_file['second_y'] = second_score.y
```

```python
submision_file['third_floor'] = third_score.floor
submision_file['third_x'] = third_score.x
submision_file['third_y'] = third_score.y
```

```python
submision_file['fourth_floor'] = fourth_score.floor
submision_file['fourth_x'] =fourth_score.x
submision_file['fourth_y'] = fourth_score.y
```

```python
submision_file['fifth_floor'] = fifth_score.floor
submision_file['fifth_x'] = fifth_score.x
submision_file['fifth_y'] = fifth_score.y
```

```python
submision_file['six_floor'] = six_score.floor
submision_file['six_x'] = six_score.x
submision_file['six_y'] = six_score.y
```

```python
submision_file.sample(10)
```

```python
submision_file['mean_x'] = submision_file[['first_x','second_x','third_x','fourth_x','fifth_x','six_x']].mean(axis=1)
submision_file['mean_y'] = submision_file[['first_y','second_y','third_y','fourth_y','fifth_y','six_y']].mean(axis=1) # ,'fourth_y'
submision_file['frequent_floor'] = submision_file[['first_floor']] # ,'third_floor','fourth_floor'
```

```python
submision_file.sample(10)
```

```python
final_subm = submision_file[['site_path_timestamp','frequent_floor','mean_x','mean_y']]
final_subm.columns = ['site_path_timestamp', 'floor', 'x','y']
final_subm.sample(6)
final_subm.to_csv('submission_best.csv', index=False)
```
